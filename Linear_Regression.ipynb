```python
from pydataset import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
%matplotlib inline
```

```python
pima = data('Pima.tr')
```

```python
pima.plot(kind='scatter' , x='skin' , y = 'bmi')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
```

```python
plt.scatter(x_train, y_train, label='training data', color='red')
plt.scatter(x_test, y_test, label='testing data', color='blue')
plt.legend()

plt.show
```

```python
LR = LinearRegression()
LR.fit(x_train.values.reshape(-1,1), y_train.values)
```

```python
prediction = LR.predict(x_test.values.reshape(-1,1))
plt.plot(x_test, prediction, color='green', linewidth=3, label= 'Lineal Regression')
#plt.scatter(x_train, y_train, label='training data', color='red')
plt.scatter(x_test, y_test, label='testing data', color='blue')
plt.legend()

plt.show
```

```python
LR.predict(np.array([[50]]))[0]
```

```python
LR.score(x_train.values.reshape(-1,1), y_train.values)
```
